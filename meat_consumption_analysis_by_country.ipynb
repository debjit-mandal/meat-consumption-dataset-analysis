# Meat Consumption Analysis by Country
This notebook provides a detailed analysis of meat consumption trends by country and meat type.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/world_meat_consumpiton_trends.csv')

# Data Preprocessing
df.dropna(subset=['Value'], inplace=True)
df['TIME'] = pd.to_datetime(df['TIME'], format='%Y')

# List of unique countries in the dataset
countries = df['LOCATION'].unique()
countries
```

## Meat Consumption Trends by Country

```python
# Meat consumption trends for each country plot
for country in countries:
    plt.figure(figsize=(14, 7))
    country_data = df[df['LOCATION'] == country]
    sns.lineplot(data=country_data, x='TIME', y='Value', hue='SUBJECT')
    plt.title(f'Meat Consumption Trends in {country}')
    plt.xlabel('Year')
    plt.ylabel('Consumption')
    plt.legend(title='Type of Meat')
    plt.xticks(rotation=45)
    plt.show()
```
