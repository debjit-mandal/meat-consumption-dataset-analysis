# Predictive Modeling of Meat Consumption Trends
This notebook provides predictive modeling for forecasting future meat consumption trends using time series analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA

df = pd.read_csv('data/world_meat_consumpiton_trends.csv')

# Data Preprocessing
df.dropna(subset=['Value'], inplace=True)
df['TIME'] = pd.to_datetime(df['TIME'], format='%Y')

# Time Series Analysis for Beef Consumption in Australia
country = 'AUS'
meat_type = 'BEEF'
country_meat_data = df[(df['LOCATION'] == country) & (df['SUBJECT'] == meat_type)]
country_meat_data.set_index('TIME', inplace=True)

# Historical data plot
plt.figure(figsize=(14, 7))
plt.plot(country_meat_data['Value'], label='Historical')
plt.title(f'{meat_type} Consumption in {country}')
plt.xlabel('Year')
plt.ylabel('Consumption')
plt.legend()
plt.show()

# Build ARIMA model
model = ARIMA(country_meat_data['Value'], order=(1, 1, 1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=10)

# Forecast plot
plt.figure(figsize=(14, 7))
plt.plot(country_meat_data['Value'], label='Historical')
plt.plot(pd.date_range(start=country_meat_data.index.max(), periods=10, freq='A'), forecast, label='Forecast')
plt.title(f'{meat_type} Consumption Forecast in {country}')
plt.xlabel('Year')
plt.ylabel('Consumption')
plt.legend()
plt.show()
```

## Predictive Modeling for Other Countries and Meat Types

```python
# Forecast for any country and meat type
def forecast_meat_consumption(country, meat_type):
    country_meat_data = df[(df['LOCATION'] == country) & (df['SUBJECT'] == meat_type)]
    country_meat_data.set_index('TIME', inplace=True)
    
    # Historical data plot
    plt.figure(figsize=(14, 7))
    plt.plot(country_meat_data['Value'], label='Historical')
    plt.title(f'{meat_type} Consumption in {country}')
    plt.xlabel('Year')
    plt.ylabel('Consumption')
    plt.legend()
    plt.show()
    
    # ARIMA model
    model = ARIMA(country_meat_data['Value'], order=(1, 1, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=10)
    
    # Forecast plot
    plt.figure(figsize=(14, 7))
    plt.plot(country_meat_data['Value'], label='Historical')
    plt.plot(pd.date_range(start=country_meat_data.index.max(), periods=10, freq='A'), forecast, label='Forecast')
    plt.title(f'{meat_type} Consumption Forecast in {country}')
    plt.xlabel('Year')
    plt.ylabel('Consumption')
    plt.legend()
    plt.show()
    
    return model_fit.summary()
```

```python
forecast_meat_consumption('USA', 'PIG')
forecast_meat_consumption('IND', 'POULTRY')
```
